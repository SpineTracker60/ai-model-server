```python
import time
from selenium import webdriver
from selenium.webdriver import Keys
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
```

```python
chrome_option = webdriver.ChromeOptions()
chrome_option.add_experimental_option('detach',True)
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()),options=chrome_option)
```

```python
max_review_count = 10000

driver.maximize_window()
url = 'https://play.google.com/store/apps/details?id=cc.forestapp&hl=ko-KR'
driver.get(url)
time.sleep(2)
```

```python
driver.find_element(By.XPATH,'//*[@id="yDmH0d"]/c-wiz[2]/div/div/div[2]/div[2]/div/div[1]/div[1]'
                    '/c-wiz[4]/section/header/div/div[2]/button').click()

time.sleep(1)

page = driver.find_element(By.CLASS_NAME,'fysCi')
page.click()
```

```python
pre_count, new_count = 0,0

while True:
    # print("pre_count : ",pre_count)
    pre_count = new_count

    driver.find_element(By.TAG_NAME,'body').send_keys(Keys.END)

    time.sleep(3)

    # 새로 불러온 댓글 수
    new_count = len(driver.find_elements(By.CLASS_NAME,'RHo1pe'))

    # total_count += new_count

    print("pre_count : ",pre_count)
    print("new count : ",new_count)

    if pre_count == new_count:
        break
    elif new_count > max_review_count:
        break

    # if new_count > max_review_count:
    #     break

elements = driver.find_elements(By.CLASS_NAME,'RHo1pe')[:max_review_count]


result_review = []
result_stars = []


for item in elements:
    result_review.append(item.find_element(By.CLASS_NAME,'h3YV2d').text)
    result_stars.append(item.find_element(By.CLASS_NAME,'iXRFPc').get_attribute('aria-label'))
    # result_stars.append(item.find_element(By.CLASS_NAME,'Jx4nYe').get_attribute('aria-label'))
    # result_stars.append(item.get_attribute("aria-label"))
```

```python
from konlpy.tag import Okt
from collections import Counter
```

```python
text_data = ''.join(list(df['comment']))
text_data
```

```python
okt = Okt()
nouns = okt.nouns(text_data)
print(nouns)
```

```python
#빈도수 체크
counter = Counter(nouns)
print(counter)
```

```python
noun_data = Counter({x:counter[x] for x in counter if len(x) > 1})
noun_data.most_common(10)
```

```python
import re
from tqdm import tqdm

noun_result = []  #[ [] , [], [] ]

for temp in tqdm(df['comment']):
    han = re.compile('[^ㄱ-ㅣ가-힣]')
    temp_result = han.sub(' ',temp) #문장에서 한글만 남기고 제거
    temp_noun = okt.nouns(temp_result)
    result_noun = [ x for x in temp_noun if len(x) >1]
    noun_result.append(result_noun)

print(noun_result)
```

```python
# bag of word 벡터 생성
from sklearn.feature_extraction.text import CountVectorizer

count_vectorizer = CountVectorizer(tokenizer=lambda x:x,lowercase=False)
bow_vect = count_vectorizer.fit_transform(noun_result)
word_list = count_vectorizer.get_feature_names_out()
print(word_list)
print(len(word_list))
```

```python
# if-idf 생성
from sklearn.feature_extraction.text import TfidfTransformer

tf_vectorizer = TfidfTransformer()
# bow vector를 tf_vectorizer에 넣으면 됌
tf_idf = tf_vectorizer.fit_transform(bow_vect)
# 단어의 가중치가 계산됌
tf_idf.toarray()
```

```python
import numpy as np
df['score_real'] = np.where(df['score']>= 4,1,0)
df.head(5)
```

```python
df['score_real'].value_counts()
```

### 모델 학습

```python
from sklearn.model_selection import train_test_split

x = tf_idf
y = df['score_real']

# 데이터 불균형 => stratify
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=1,stratify=df['score_real'])
```

```python
X_train.shape , X_test.shape
```

```python
# 모델 학습 (이진 분류)

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score

lr_model = LogisticRegression(random_state=1)
lr_model.fit(X_train,y_train)
```

```python
y_pred = lr_model.predict(X_test)
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
precision_score(y_test,y_pred)
```

```python
recall_score(y_test,y_pred)
```

```python
f1_score(y_test,y_pred)
```

### 모델 테스트

```python
def comment_predict(text):
  sentence_test = text
  # han = re.compile('[^ㄱ- | 가-힣]')
  han = re.compile('[^ㄱ-ㅣ 가-힣]')
  temp_result = han.sub(' ',sentence_test)
  temp_noun = okt.nouns(temp_result)
  result_noun = [x for x in temp_noun if len(x)>1]

  # transform만 (fit하면 안됌)
  vec_test = count_vectorizer.transform([result_noun])
  vec_test.toarray().sum()
  tf_test = tf_vectorizer.transform(vec_test)
  tf_test.toarray().sum()

  result = ''
  if lr_model.predict(tf_test) == 1:
    result = '긍정'
  else:
    result = '부정'

  return result

print(comment_predict('역시 힐링 갓겜!'))
```
